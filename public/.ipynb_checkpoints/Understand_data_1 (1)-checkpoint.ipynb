```python
import random
res = pt_1.sample(frac = 0.1)
res.shape

from  collections import defaultdict
tup_dic = defaultdict(tuple)

def randomize(df):
    df['Sympton Tuples'] = ''
    df['Disease Tuples'] = ''
    df['FC Tuples'] = ''

    arr =  [(7,1),(5,3),(15,1),(6,2),(2,1),(1,1),(35,3)]
    arr_index = 0
    for key, item in final_dict.items():
        curr_fc = key
        for idx in numberofslots(arr[arr_index][1],arr[arr_index][0], df.shape[0]-1)          :
            dis_index = random.randint(0,len(item)-1)
            curr_disease = item[dis_index]
            curr_symptoms = disease_dict[curr_disease]
            final_tuple = (curr_fc,curr_disease,random.choice(curr_symptoms))
#             low = random.randint(0, df.shape[0]-1)
#             high = random.randint(low, df.shape[0]-1)
#             idx = random.randint(low, high) % df.shape[0]
#             print(idx)
            tup_dic[idx]= final_tuple
            
#         for temp in item:
#             curr_disease = temp
#             curr_symptoms = disease_dict[curr_disease]
#               #for symptom in curr_symptoms:
#             final_tuple = (curr_fc,curr_disease,curr_symptoms)
        
# #             count = random.randint(1,arr[arr_index])
#             for i in range(arr[arr_index]):
#                 idx = random.randint(0,df.shape[0]-1)
#                 print(idx)
#                 tup_dic[idx].append(final_tuple)
        arr_index += 1
    #print("Tuples : ",tup_dic)          
    for index, val in tup_dic.items():
    #print(val)
        fc_set, dis_set  = set(), set()
        for tup in val:
            fc_set.add(tup[0])
            dis_set.add(tup[1])

        temp = str(val)
        df.at[index,'Sympton Tuples'] = val[2]
        # temp = str(fc_set)
        df.at[index,'FC Tuples'] = " ".join(list(fc_set))
        df.at[index,'Disease Tuples'] = " ".join(list(dis_set)).replace("'", "")

  # print("Tuples : ",df['Tuples'][0])
rec_set = set()
def numberofslots(slot, number_of_rec, max_val):
    
    rem = number_of_rec
    res = []
    flag = False
    num =   slot * 2 + 1
    for i in range(num):
      
        if flag:        
            rec = int(number_of_rec / slot)
             
            rem -= rec
            lst = []

            for cnt in range(rec):
                max_range_val = min(round((i+1) * max_val/num), max_val)
                index_value = random.randint(round(i*max_val/num), max_range_val)
                while index_value in rec_set:
                    index_value = random.randint(round(i*max_val/num), max_range_val)
                rec_set.add(index_value)
                lst.append(index_value)
                lst.sort()
            res.append(lst)
        flag = not flag
    final_index = [x for item in res for x in item]
    
    print(res)
    return final_index
    
#numberofslots(1, 1, 1444)            
randomize(pt_1)
```
